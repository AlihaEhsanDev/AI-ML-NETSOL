## 📝 Task 2 — Grouped Bar Chart with Matplotlib

### Background
A **grouped bar chart** places multiple bars side by side for the same category, making it easy to compare values within each group.

### 🎯 Your Task
The data below shows **exam scores** for three subjects across two classes.

| Subject | Class A | Class B |
|---------|---------|---------|
| Math | 72 | 68 |
| Science | 85 | 80 |
| English | 78 | 74 |
| History | 65 | 70 |

Create a **grouped bar chart** with:
- Two bars (Class A and Class B) side by side for each subject
- Different colors for each class
- Value labels on top of each bar
- A legend, axis labels, and a title

> 💡 **Hint:** Use `np.arange(len(subjects))` for x-positions, then offset each group with `x - 0.2` and `x + 0.2` and `width=0.4`.  
> To add labels: loop over bars using `bar.get_height()` and `plt.text()`.


````python
subjects = ['Math', 'Science', 'English', 'History']
class_a  = [72, 85, 78, 65]
class_b  = [68, 80, 74, 70]

# YOUR CODE HERE
# Create x positions using np.arange
x=np.arange(len(subjects))
width=0.4

# Plot Class A bars (offset left)
bars1 = plt.bar(x - 0.2, class_a, width, label="Class A", color="skyblue")

# Plot Class B bars (offset right)
bars2 = plt.bar(x + 0.2, class_b, width, label="Class B", color="orange")

# Add value labels above each bar
for bar in bars1:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, str(height),
             ha='center', va='bottom')

for bar in bars2:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, str(height),
             ha='center', va='bottom')

# Add title, axis labels, legend, and x-tick labels
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.title("Exam Scores Comparison: Class A vs Class B")
plt.xticks(x, subjects)
plt.legend()

plt.show()
````

## 📝 Task 7 — Interactive Line Plot with Bokeh

### Background
Bokeh creates **interactive, browser-rendered** plots. The `HoverTool` lets users see exact data values when they hover over points.

### 🎯 Your Task
Create an **interactive Bokeh line plot** showing website traffic over 7 days:

```
days    = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
visitors = [120, 180, 150, 200, 250, 300, 270]
```

Requirements:
- Line width = 2, color of your choice
- Circle markers on each data point (size = 8)
- A `HoverTool` that shows the day and visitor count on hover
- Proper title and axis labels

> 💡 **Hint:** To use HoverTool with plain lists, convert data to a `ColumnDataSource`.  
> `source = ColumnDataSource(data=dict(days=days, visitors=visitors))`  
> Then use `p.line(x='days', y='visitors', source=source)` and  
> `HoverTool(tooltips=[("Day", "@days"), ("Visitors", "@visitors")])`.


````python
output_notebook()

days     = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
visitors = [120, 180, 150, 200, 250, 300, 270]

# YOUR CODE HERE
# 1. Create a ColumnDataSource
source = ColumnDataSource(data=dict(days=days, visitors=visitors))

# 2. Create a Bokeh figure with title and axis labels
p = figure(title="Website Traffic Over 7 Days",
           x_range=days,
           x_axis_label='Day',
           y_axis_label='Visitors')


# 3. Add a line renderer
p.line('days','visitors', source=source, line_width=2, color='blue')


# 4. Add circle markers
p.circle('days','visitors', source=source, size=8)


# 5. Add a HoverTool
hover = HoverTool(tooltips=[("Day","@days"), ("Visitors","@visitors")])
p.add_tools(hover)


# 6. Show the plot
show(p)
````

## 📝 Task 9 — Choose Your Own Visualization

### Background
Good data visualization starts with choosing the **right chart type** for your data and message. This task tests that judgment.

### 🎯 Your Task
Below is a dataset of a fictional **e-commerce store's weekly performance**:

```python
data = {
    'Day':        ['Mon','Tue','Wed','Thu','Fri','Sat','Sun'],
    'Orders':     [45,   62,   58,   70,   95,   110,  80 ],
    'Revenue':    [900,  1240, 1160, 1400, 1900, 2200, 1600],
    'Returns':    [2,    3,    5,    4,    6,    8,    5   ],
    'Category':   ['Electronics','Clothing','Books','Electronics','Clothing','Books','Electronics']
}
```

**You must create exactly 3 different chart types** using any combination of Matplotlib, Seaborn, or Plotly. For each chart:
1. Clearly state (in a comment) **which chart type you chose and why** it suits that particular relationship
2. Include a title, axis labels, and any useful annotations

**Suggested angles to explore** (pick any 3):
- Trend of `Orders` or `Revenue` over the week
- Proportion of `Orders` by `Category`
- Relationship between `Orders` and `Revenue`
- Comparison of `Returns` vs `Orders` by day

> 💡 There is no single correct answer here. Focus on making each chart **clear and informative**.


````python
df = pd.DataFrame({
    'Day':      ['Mon','Tue','Wed','Thu','Fri','Sat','Sun'],
    'Orders':   [45,   62,   58,   70,   95,  110,   80 ],
    'Revenue':  [900, 1240, 1160, 1400, 1900, 2200, 1600],
    'Returns':  [2,    3,    5,    4,    6,    8,    5   ],
    'Category': ['Electronics','Clothing','Books','Electronics','Clothing','Books','Electronics']
})

# ── Chart 1 ──────────────────────────────────────────────────────────────
# Chart type chosen: Line chart
# Why it suits this data: A line chart is ideal for showing trends over time. Here, we can see how Orders and Revenue change across the week.

plt.figure(figsize=(10,5))
plt.plot(df['Day'], df['Orders'], marker='o', label='Orders', color='blue')
plt.plot(df['Day'], df['Revenue'], marker='s', label='Revenue', color='green')
plt.title('Orders and Revenue Trend Over the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Count / Revenue ($)')
plt.legend()
plt.grid(True)

# Annotate peaks
for i, val in enumerate(df['Orders']):
    plt.text(i, val+2, str(val), ha='center', color='blue')
for i, val in enumerate(df['Revenue']):
    plt.text(i, val+20, str(val), ha='center', color='green')

plt.show()
````

````python
# ── Chart 2 ──────────────────────────────────────────────────────────────
# Chart type chosen: Pie chart
# Why it suits this data: A pie chart shows proportion. We can visualize the proportion of Orders coming from each Category.

category_orders = df.groupby('Category')['Orders'].sum()

# Create pie chart
plt.figure(figsize=(7,5))
plt.pie(category_orders,
        labels=category_orders.index,
        autopct='%1.1f%%',
        startangle=140,
        colors=['#ff9999','#66b3ff','#99ff99'])

plt.title('Proportion of Orders by Category')

plt.show()

````

````python
# ── Chart 3 ──────────────────────────────────────────────────────────────
# Chart type chosen: Scatter plot with regression line
# Why it suits this data: Scatter plots are great to show relationships between two numeric variables. Adding a regression line shows correlation between Orders and Revenue.
# Create DataFrame
sns.set(style="whitegrid")
plt.figure(figsize=(8,5))
sns.regplot(x='Orders', y='Revenue', data=df, scatter_kws={'s':80, 'color':'purple'}, line_kws={'color':'orange'})
plt.title('Relationship Between Orders and Revenue')
plt.xlabel('Orders')
plt.ylabel('Revenue ($)')

# Annotate points
for i in range(len(df)):
    plt.text(df['Orders'][i]+1, df['Revenue'][i]+10, df['Day'][i], fontsize=9)

plt.show()
````
